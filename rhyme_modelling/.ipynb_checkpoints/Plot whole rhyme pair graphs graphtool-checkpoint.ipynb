```python
from graph_tool import Graph
from graph_tool.draw import graph_draw
from graph_tool.topology import extract_largest_component

import pandas as pd
```

## Plot rhyme pair graph pre clustering

```python
df = pd.read_csv("tsvs/positive_pairs.tsv", sep="\t")
df
```

```python
pairs = set(zip(*(df.word_a, df.word_b)))
len(pairs)
```

```python
vocab = set(df.word_a)
vocab.update(set(df.word_b))
len(vocab)
```

```python
edges = pairs.copy()
vertices = list(vocab)

G = Graph(directed=False)

vertices_ = G.add_vertex(len(vertices))

vertex_lookup = {v: v_ for v, v_ in zip(vertices, vertices_)}
vertex_lookup

for w1, w2 in edges:
    G.add_edge(vertex_lookup[w1], vertex_lookup[w2])
```

```python
graph_draw(G)
```
