```python
import tarfile
import re
from pathlib import Path
pattern = r".*[dD]ikt.*|.*[lL]yrik.*|.*[Ss]alme.*|.*[Ss]ang.*"

p = Path("dikt_split/bokmål")
pp = Path("dikt_split/nynorsk")


tar = tarfile.open('20150526_nbdig_txt01.tar.gz')

i = 0

for file in tar:
    if file.isfile():
        if re.match(pattern, file.name):
            n = file.name.split("digibok_")[1]
            year = n.split("-")[1]
            lang = n.split("-")[2]
            if year:
                year = int(year)
                if year <= 1950 and year >= 1920:
                    if lang == "nob":
                        fp = p / n
                    if lang == "nno":
                        fp = pp / n
                    if fp:
                        i+=1
                        f = tar.extractfile(file)
                        content = f.read()

                        with fp.open("wb") as f:
                            f.write(content)
                            
print(i)
        
```

## Flytt dikt_split til poems (og tell linjer)

```python
from pathlib import Path
np = Path("dikt_split/bokmål")

bp = Path("poems/bokmål")
nnp = Path("poems/nynorsk")


dikt_start = "#---START---#"
dikt_slutt = "#---SLUTT---#"
alle = []
linjer = 0

for e in np.iterdir():
    if e.is_file():
        with e.open() as f:
            content = f.read()
            s = content.split(dikt_start)
            diktene = [d.split(dikt_slutt)[0] for d in s if len(d.split(dikt_slutt))>1]
            n = e.name[:-4]
            print(n, len(diktene))
            for i, dikt in enumerate(diktene):
                if dikt.startswith("\nNYNORSK"):
                    dikt = dikt[8:]
                    fp = nnp / f"{n}_{i+1}.txt"
                else:
                    fp = bp / f"{n}_{i+1}.txt"

                with fp.open("w+") as f:
                    f.write(dikt[1:])

                bitene = dikt.split("\n\n")
                linjene = [bit.split("\n") for bit in bitene]
                for linje in linjene:
                    linjer += len(linje)
print(linjer)
```

## Flytt diktbitene til ny mappe

```python
from pathlib import Path

p = Path("dikt_split")
np = Path("poems")

dikt_start = "#---START---#"
dikt_slutt = "#---SLUTT---#"

alle_linjene = 0
ler = {}
for x in p.iterdir():
    with x.open() as f:
        content = f.read()
        if content.startswith("dårlig kvalitet"):
            continue
        s = content.split(dikt_start)
        diktene = [d.split(dikt_slutt)[0].strip() for d in s if len(d.split(dikt_slutt))>1]
        for i, dikt in enumerate(diktene):
            name = x.name.split("_")[0]
            newfile = np / f"{name}_{i+1}.txt"
            with open(newfile, "w+") as f:
                    f.write(dikt)
```

## Telle antall characters i linjene

```python
from pathlib import Path
import numpy as np

p = Path("poems")
m = 0
ll = []
for poem in p.iterdir():
    with poem.open("r") as f:
        lines = f.readlines()
    ll += [len(l) for l in lines]
    mm = max([len(l) for l in lines])
    if mm > m:
        m = mm
    
```

## Make subset for inter-annotator agreement

```python
from pathlib import Path
from random import randint

p = Path("poems")
l = list(p.iterdir())
miniset = []
tot_stanzas = 0

while tot_stanzas < 75:
    i = randint(0, len(l))
    pp = l.pop(i)
    with pp.open("r") as f:
        content = f.read()
    stanzas = content.split("\n\n")
    tot_stanzas += len(stanzas)
    miniset.append(pp.name)
miniset
```

## Telle år og sortere på år

```python
from pathlib import Path
import numpy as np

p = Path("dikt-idetfri")
l = [n.name for n in p.iterdir()]
print(f"Length old ocr: {len(l)}")

years_str = [n.split("digibok")[1].split("-")[1] for n in l]
y = [int(y) if y else 0 for y in years_str]
years = [int(y) for y in years_str if y]

name_y_old = list(zip(l, y))
name_y_old.sort(key=lambda x: x[1])
_, y = zip(*name_y_old) 

i = y.index(1900)
newer_books = name_y_old[i:]

print(f"""
GAMMEL OCR
Median: {np.median(years)}
Mean: {np.mean(years)}
Max: {np.max(years)}
Min: {np.min(years)}
No year: {len(years_str)-len(years)}
""")



p = Path("sammenlikn-ocr")
l = []
years2 = []
y = []

for d in p.iterdir():
    for f in d.iterdir():
        if f.name.startswith("digibok"):
            l.append(f.name)
            year = f.name.split("digibok")[1].split("-")[1]
            if year:
                years2.append(int(year))
                y.append(int(year))
            else:
                y.append(0)
                
name_y_new = list(zip(l, y))
name_y_new.sort(key=lambda x: x[1])

print(f"Length new ocr: {len(l)}")

print(f"""
NY OCR
Median: {np.median(years2)}
Mean: {np.mean(years2)}
Max: {np.max(years2)}
Min: {np.min(years2)}
No year: {len(y)-len(years2)}
""")
```

```python
from pathlib import Path

s1 = set(newer_books)
s2 = set(name_y_new)

get_new_ocr = s1 - (s1.intersection(s2))
print(len(get_new_ocr))
print(len(s1))

codes = [name.split("digibok_")[1].split("-")[0] for name, year in get_new_ocr]

p = Path("diktkoder.txt") 
p.write_text("\n".join(codes))

get_new_ocr

_, years3 = zip(*get_new_ocr)

print(f"Length potential new ocr: {len(years3)}")

print(f"""
NY NY OCR
Median: {np.median(years3)}
Mean: {np.mean(years3)}
Max: {np.max(years3)}
Min: {np.min(years3)}
""")


potential_years = years2 + list(years3)

print(f"Length total new ocr: {len(potential_years)}")

print(f"""
NY + NY NY OCR
Median: {np.median(potential_years)}
Mean: {np.mean(potential_years)}
Max: {np.max(potential_years)}
Min: {np.min(potential_years)}
""")
```
